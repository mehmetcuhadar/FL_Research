```python
import pandas as pd
import numpy as np
from sklearn import linear_model, svm
from sklearn.cluster import KMeans

def read_label_flipping_data():
    df = pd.read_csv("3001_results.csv", header= None)
    return df

def read_backdoor_data():
    df = pd.read_csv("backdoor_results.csv",header= None)
    return df

def read_workers_selected_data():
    df = pd.read_csv("3003_workers_selected.csv", header= None)
    return df

def read_poisoned_workers():
    file1 = open("logs/3003.log", 'r')
    Lines = file1.readlines()
    w_arr = []
    start = 0
    end = 0
    
    for line in Lines:
        if "Poisoning data for workers:" in line:
            for i in range(len(line)):
                if line[i] == "[":
                    start = i
                if line[i] == "]":
                    end = i
            w_arr = line[start+1:end]
    w_arr = w_arr.replace(" ", "").split(',')
    poisoned_workers = []
    for i in range(len(w_arr)):
        poisoned_workers.append(int(w_arr[i]))
    return poisoned_workers

def get_workers_binary_array(workers):
    arr = []
    for t in range(workers.shape[0]):
        label = []
        for i in range(50):
            if i in workers[t]:
                label.append(1)
            else:
                label.append(0)
        arr.append(label)
    return arr

def get_prediction_workers_binary_array():
    arr2 = []
    for i in range(50):
        label = []
        for j in range(50):
            if i == j:
                label.append(1)
            else:
                label.append(0)
        arr2.append(label)
    return arr2

def get_malicious_clients(y_kmeans):
    arr1 = []
    arr2 = []
    arr3 = []
    for i in range(50):
        if y_kmeans[i] == 0:
            arr1.append(i)
        elif y_kmeans[i] == 1:
            arr2.append(i)
        else:
            arr3.append(i)
    print("---",arr1)
    print("---",arr2)
    print("---",arr3)
    if len(arr1)>len(arr2):
        return arr2
    else:
        return arr1

if __name__ == '__main__':

    datas = [(2*A/(2*A+C+B)), (2*A/(2*A+B)), (2*A/(2*A+C)), (A/(A+B)), (A/(A+B+C)), (A/(B+C)), (A/(A+C)), df[0]]
    datas = [B/(A+B), B/(A+B+C), (A+B)/(A+B+C)]
    datas = [(A+B)/(A+C)]
    datas = [B/(A+B), df[0]]
    
    for data in datas:
        workers = np.array(read_workers_selected_data())
        X = get_workers_binary_array(workers)
        target = 0
        y = np.array(data[13 + target])*100  
        lm = svm.SVR(gamma = "auto",kernel = "poly", coef0 = 10)
        data = (data - data.min()) / (data.max() - data.min())
        data *= 100/data.max()
        model = lm.fit(X[100:], data[100:])
        p_bin_arr = get_prediction_workers_binary_array()
        predictions = lm.predict(p_bin_arr)
        predictions = (predictions - predictions.min()) / (predictions.max() - predictions.min())
        #print(predictions[predictions < 0.5])
        print(np.where(predictions < 0.5))
        plt.plot(predictions)
        print(np.argsort(predictions,0))
        kmeans = KMeans(n_clusters=4)
        
        y_kmeans = kmeans.fit_predict(predictions.reshape(50,1))
        malicious_clients = get_malicious_clients(y_kmeans)
        print(read_poisoned_workers())
        print( malicious_clients)
```
